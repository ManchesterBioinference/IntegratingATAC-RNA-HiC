```r
library(tidyverse)
```

```r
data_gene <- data.frame(x=c('Early','Intermediate I','Intermediate II', 'Late','Repressed'), 
                        y=c(0.437337543640406, 0.626790692672614, 0.485188099454553, 0.637875603148071, 0.281999325158577))
data_LADS <- data.frame(x=c('LAD1','LAD2','LAD3'), y=c(0.591557381578406, 0.393839268757538, 0.386124061754459))
```

```r
p1 <- ggplot(data=data_LADS, aes(x=x,y=y)) + geom_col(fill='yellow2', width=0.5) + 
xlab('') + 
ylab('LFC') +
ggtitle('LFC of LADs in Changing A/B Compartments')+
theme(panel.background=element_blank(),plot.title = element_text(hjust = 0.5, face='bold', size=16),
         axis.title=element_text(size=16,face="bold"), axis.text.x=element_text(size=14, face='bold', hjust=0.5), 
         axis.text.y=element_text(size=14, face='bold'),axis.line=element_line('black'),strip.text.x=element_blank(), strip.text.y=element_blank(), legend.position="none") 
p1
ggsave('../../../IntegratingATAC-RNA-HiC/Individual_Figures_Files/Supp_fig7/LADS.pdf',p1)
```

```r
p2 <- ggplot(data=data_gene, aes(x=x,y=y)) + geom_col(fill='coral1', width=0.6) + 
xlab('') + ylab('LFC') + 
ggtitle('LFC of Genes in Changing A/B Compartments')+
theme(panel.background=element_blank(),plot.title = element_text(hjust = 0.5, face='bold', size=16),
         axis.title=element_text(size=16,face="bold"), axis.text.x=element_text(size=14, face='bold', hjust=1.0,  angle=25), 
         axis.text.y=element_text(size=14, face='bold'),axis.line=element_line('black'),strip.text.x=element_blank(), strip.text.y=element_blank(), legend.position="none") 
p2
ggsave('../../../IntegratingATAC-RNA-HiC/Individual_Figures_Files/Supp_fig7/gene.pdf',p2)
```
