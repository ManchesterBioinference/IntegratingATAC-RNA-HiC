```r
tmp_list <- read.table('tmp_list_unique.txt', header=T)
head(tmp_list)
for (ii in 32:dim(tmp_list)[1]){
    
data_tmp <- subset(data, data$ENSG %in% tmp_list$ENSG[ii])
#print(head(data_tmp))
if (dim(data_tmp)[1]!=0){
gene_name <- tmp_list$GENE_NAME[ii]

tmp_atacseq <- unique(data_tmp[,c('AT0','AT20','AT1H','AT2H','AT4H','AT24H')])
names(tmp_atacseq) <- c('T0','T20','T1H','T2H','T4H','T24H')
tmp_atacseq$label <- paste('atacseq',1:dim(tmp_atacseq)[1],sep='')
melted_tmp_atacseq <- melt(tmp_atacseq)

tmp_chic <- unique(data_tmp[,c('T0','T20','T1H','T4H','T24H')])
tmp_chic$label <- paste('CHiC', 1:dim(tmp_chic)[1],sep='')
melted_tmp_chic <- melt(tmp_chic)

tmp_gene <- unique(data_tmp[,c('ET0_a','ET20_a','ET1H_a','ET2H_a','ET4H_a','ET24H_a')])
names(tmp_gene) <- c('T0','T20','T1H','T2H','T4H','T24H')
tmp_gene$label <- c('gene')
melted_tmp_gene <- melt(tmp_gene)

tmp_corr <- data_tmp[,c('corr_atacseq_gene', 'corr_atacseq_otherEnd','corr_gene_otherEnd')]
names(tmp_corr) <- c('atac_gene','atac_chic','gene_chic')
tmp_corr$label <- 1:dim(tmp_corr)[1]
melted_tmp_corr <- melt(tmp_corr,id='label')
melted_tmp_corr$label2 <- 'correlation'

head(melted_tmp_atacseq)
head(melted_tmp_chic)
head(melted_tmp_gene)
melted_tmp_atacseq$label2 <- 'atacseq'
melted_tmp_chic$label2 <- 'chic'
melted_tmp_gene$label2 <- 'gene'

melted_tmp <- rbind(melted_tmp_atacseq, melted_tmp_chic, melted_tmp_gene)

p1 <- ggplot() + geom_line(data=melted_tmp, aes(x=variable, y=value, group=label,color=label))  + 
#geom_text(data=melted_tmp, aes(x=))
geom_tile(data=melted_tmp_corr, aes(y=as.numeric(label), x=variable, fill=value)) +
scale_fill_gradient2(low = "blue", mid = "white",
  high = "red") +
xlab('Time') + ylab('Expression') + ggtitle(gene_name) + facet_wrap( label2~., scales = "free",ncol=2)
    
ggsave(paste('tmp/',gene_name,'.png',sep=''), p1)}
}
```
