### load packages required

```r
library(ggplot2)
```

### load the interaction matrices. Only interaction matrices for Chr1 with resolution of 100 kb are illustrated here

```r
data_T0 <- read.table('https://personalpages.manchester.ac.uk/staff/jing.yang/Data/HiC/data/T0_chr1_100kresolution.txt', header=T, row.names=1)
data_T20 <- read.table('https://personalpages.manchester.ac.uk/staff/jing.yang/Data/HiC/data/T20_chr1_100kresolution.txt', header=T, row.names=1)
data_T1H <- read.table('https://personalpages.manchester.ac.uk/staff/jing.yang/Data/HiC/data/T1H_chr1_100kresolution.txt', header=T, row.names=1)
data_T4H <- read.table('https://personalpages.manchester.ac.uk/staff/jing.yang/Data/HiC/data/T4H_chr1_100kresolution.txt', header=T, row.names=1)
data_T24H <- read.table('https://personalpages.manchester.ac.uk/staff/jing.yang/Data/HiC/data/T24H_chr1_100kresolution.txt', header=T, row.names=1)
```

### plot interaction matrices for chr1 at time 0 mins, 20 mins, 1 hr, 4hrs and 24 hrs

```r
col=colorRampPalette(c("yellow", "red"))(20) 
M <- as.matrix(log(data_T0[,2:dim(data_T0)[2]]))
#M[lower.tri(M)] <- NA
image(M,col = col,xaxt="n",yaxt="n",xlab="Chr1",ylab="Chr1", font.main = 3)
title(main = "Interaction matric at time 0 mins", font.main = 4)
```

```r
M <- as.matrix(log(data_T20[,2:dim(data_T0)[2]]))
#M[lower.tri(M)] <- NA
image(M,col = col,xaxt="n",yaxt="n",xlab="Chr1",ylab="Chr1", font.main = 3)
title(main = "Interaction matrix at time 20 mins", font.main = 4)
```

```r
M <- as.matrix(log(data_T1H[,2:dim(data_T0)[2]]))
#M[lower.tri(M)] <- NA
image(M,col = col,xaxt="n",yaxt="n",xlab="Chr1",ylab="Chr1", font.main = 3)
title(main = "Interaction matrix at time 1 hr", font.main = 4)
```

```r
M <- as.matrix(log(data_T4H[,2:dim(data_T0)[2]]))
#M[lower.tri(M)] <- NA
image(M,col = col,xaxt="n",yaxt="n",xlab="Chr1",ylab="Chr1", font.main = 3)
title(main = "Interaction matrix at time 4 hrs", font.main = 4)
```

```r
M <- as.matrix(log(data_T24H[,2:dim(data_T24H)[2]]))
#M[lower.tri(M)] <- NA
image(M,col = col,xaxt="n",yaxt="n",xlab="Chr1",ylab="Chr1", font.main = 3)
title(main = "Interaction matrix at time 24 hrs", font.main = 4)
```
